```python
import sympy as sy
x, y, h = sy.symbols('x y h')
```

###**a)**
$m = PQ = \frac{y_1 - y_0}{x_1 - x_0 }$

Para calcular o coeficiente angular utilizando limites:

$\lim_{Q \to P} m = \lim_{x_1 \to x_0} \frac{y_1 - y_0}{x_1 - x_0 }$

Aplicar a fun√ß√£o, substituindo os valores de $y_1$ e $y_2$

$f(x) = x¬≤$


$m = \frac{x_1¬≤ - x_0¬≤}{x_1 - x_0} = \frac{(x_1 + x_0)(x_1 - x_0)}{x_1 - x_0} = x_1 + x_0$

No ponto em que $x_1 = x_0$, $m=2x_0$

R: $m = 2$

###**b)**
$m=2x = 2(-\frac{1}{2}) = -1$


```python
f=x**2
points = [(1,1,2),(-1/2, 1/4, -1)]
tangents = []
for (x0, y0, m) in points:
  tangent = m*(x - x0) + y0
  tangents.append(tangent)

p = sy.plot(f, tangents[0], tangents[1], (x,-2,2), show=False, legend=True)
p[0].line_color = 'blue'
p[0].label = 'f(x) = x^2'
p[1].line_color = 'green'
p[1].label = 'Tangente em (1,1)'
p[2].line_color = 'red'
p[2].label = 'Tangente em (-1/2,1/4)'
p.show()
```

### Aplicando Diferen√ßas Finitas
$m = \frac{f(x_0 + h) -y}{(x_0 + h) - x_0}$

Sabendo que $f(x) = x¬≤$, substitui $f(x_0 + h)$ por $(x_0 + h)¬≤$ e $y$ por $x¬≤$

$m = \frac{f(x_0 + h) -y}{(x_0 + h) - x_0} = \frac{(x_0 + h)¬≤ -x_0¬≤}{(x_0 + h) - x_0} = \frac{x_0¬≤ + 2x_0h +h¬≤ -x_0¬≤}{h} = \frac{2x_0h +h¬≤}{h} = 2x_0 + h$

Quando h tende a zero:

$\lim_{h \to 0} = 2x_0 + 0 = 2x_0$

[CONFIRMAR COM O PROFESSOR]

Quando $x \to 1$, substitui $x$ por 1 em $2x_0$ = 2

Quando $x \to -1/2$, substitui $x$ por -1/2 em $2x_0$ = -1

**S√£o os mesmos valores do coeficiente angular da tangente.**


```python
f = x**2
limit1 = sy.limit(f, x, 1)
limit2 = sy.limit(f, x, -1/2)

limit1, limit2
```
